```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

pd.options.mode.chained_assignment = None # hide some warnings
```

```python
# Read dog data, extract rating into new dataframe
dog_data = pd.read_csv('dog_rates_tweets.csv', parse_dates=['created_at'])
rating_data = dog_data[dog_data['text'].str.contains('\d+\/10')]
```

```python
# Extract exact rating and cast as float
rating_data['rating'] = rating_data['text'].str.extract(pat= r'(\d*?)\/10')
rating_data['rating'] = rating_data['rating'].astype(float)
```

```python
# Filter ratings that are less than or equal to 25
rating_data = rating_data[rating_data['rating'] <= 25]
rating_data.dtypes
```

```python
# Show scatterplot of results so far
plt.plot(rating_data['created_at'], rating_data['rating'], '.')
```

```python
# Extract timestamp into new column
rating_data['timestamp'] = rating_data['created_at'].apply(lambda date : date.timestamp())
```

```python
# Set up linear regression, print relevant parameters
regression = stats.linregress(rating_data['timestamp'], rating_data['rating'])
print('Regression slope: ', regression.slope)
print('Regression intercept: ', regression.intercept)
```

```python
# Set up plot
plt.title('Inflation in dog ratings over time')
plt.xlabel('Date')
plt.ylabel('Ratings')
plt.plot(rating_data['timestamp'], rating_data['rating'], '.')
plt.plot(rating_data['timestamp'], regression.slope*rating_data['timestamp'] + regression.intercept, linewidth= 3)
```

```python
# Show plot
plt.show()
```
